```python
def vectoring(encod):
  img_path="/content/drive/MyDrive/kaggle_data/transformed/"
  pif=img_path+encod
  print(pif)
  return image_encoder.encode(pif)

def change_lis(endc):
  change_typ=[float(i) for i in endc.replace('[', '').replace(']', '').split(',')]
  return change_typ

def res_image(endx):
  img_path="/content/drive/MyDrive/kaggle_data/train_images/"
  new_path="/content/drive/MyDrive/kaggle_data/transformed"
  f_img =img_path+str(endx)
  img=Image.open(f_img)
  img = img.resize((325,425))
  grey_scale=img.convert('L')
  return grey_scale.save(new_path+"/"+str(endx))
```
